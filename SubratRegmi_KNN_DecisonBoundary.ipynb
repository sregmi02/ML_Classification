### KNN Classification using scikit-learn

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.plotting import plot_decision_regions #This package is used to plot decision boundaries
```

```python
#Load the dataset
iris = load_iris()
```

```python
#Using only sepal length and sepal width
X = iris.data[:, :2]
y = iris.target
```

```python
#Plot the data
plt.figure(figsize = (7,6))
plt.scatter (
    X[:, 0],
    X[:, 1],
    c=y,
    edgecolor="black"
)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Datset")
plt.show()
```

```python
for k in [1, 3, 5, 10]: #Iterate through values of K [1,3,5,10]
    model = KNeighborsClassifier(n_neighbors = k) #KNN model
    model.fit(X, y)

    plot_decision_regions(X, y, clf=model) #Plotting the decision boundaries
    plt.xlabel('X')  
    plt.ylabel('Y')
    plt.title(f'k = {k}')
    plt.show()
```

### How Decision Boundaries Change

With lower values of k (say k = 1), the decison boundary is very precise as it only considers one nearest neighbor, which in turn leads to overfitting. 
If we keep increasing the value of k (say k = 4 or 5), the boundaries keep getting smoother and we can get better predictions.
However, if we increase the value of k too much (k=10), the boundaries get too smooth and miss out on details of the data and leads to underfitting.
